```python
import pandas as pd
import numpy as np
```

```python
fg = pd.read_csv("./csv_files/fear_greed_index.csv")
hist = pd.read_csv("./csv_files/historical_data.csv")
```

### Clean Data

```python

fg['date'] = pd.to_datetime(fg['date']).dt.date
fg = fg[['date', 'value', 'classification']]
fg.rename(columns={'value': 'fear_greed_score'}, inplace=True)


# Convert IST timestamp into datetime
hist['Timestamp IST'] = pd.to_datetime(hist['Timestamp IST'], format="%d-%m-%Y %H:%M", errors='coerce')

# Extract date for merging
hist['date'] = hist['Timestamp IST'].dt.date

# Normalize columns
hist.columns = hist.columns.str.lower().str.replace(" ", "_")
```

```python
merged = hist.merge(fg, on="date", how="left")

print("Merged shape:", merged.shape)
merged.head(10)
```

### Feature Engineering

```python
df = merged.copy()

# Notional value of trade
df['notional'] = df['execution_price'] * df['size_tokens']

# Win/loss
df['is_win'] = (df['closed_pnl'] > 0).astype(int)

# Normalized return
df['pnl_per_usd'] = df['closed_pnl'] / df['notional'].replace(0, np.nan)

# Trade direction normalized
df['direction'] = df['direction'].str.upper()
df['sentiment'] = df['classification'].str.upper()
```

### Fear vs Greed

```python
sentiment_summary = (
    df.groupby('sentiment')
    .agg(
        n_trades=('closed_pnl', 'count'),
        avg_pnl=('closed_pnl', 'mean'),
        median_pnl=('closed_pnl', 'median'),
        win_rate=('is_win', 'mean'),
        avg_notional=('notional', 'mean'),
        avg_fee=('fee', 'mean')
    )
)

sentiment_summary
```

Direction - Sentiment

```python
direction_summary = (
    df.groupby(['sentiment', 'direction'])
    .agg(
        n_trades=('closed_pnl', 'count'),
        avg_pnl=('closed_pnl', 'mean'),
        win_rate=('is_win', 'mean'),
        avg_notional=('notional', 'mean')
    )
    .reset_index()
)

direction_summary.head ()
```

### Leverage Buckets vs Sentiment

```python
bins = [0, 2, 5, 10, 20, 50, np.inf]
labels = ['<=2x', '2–5x', '5–10x', '10–20x', '20–50x', '>50x']

df['lev_bucket'] = pd.cut(df['start_position'], bins=bins, labels=labels)

leverage_summary = (
    df.groupby(['sentiment', 'lev_bucket'])
    .agg(
        n_trades=('closed_pnl', 'count'),
        avg_pnl=('closed_pnl', 'mean'),
        win_rate=('is_win', 'mean')
    )
    .reset_index()
)

leverage_summary.head()
```
